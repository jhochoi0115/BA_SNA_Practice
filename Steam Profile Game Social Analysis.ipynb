# 스팀프로그램 및 게임 장르의 SNA

  - Relation의 Direction은 필요 없음 (공존)

```r
JagaEdge %>%
  as_tbl_graph() %>%
  ggraph(layout='kk') + 
  geom_node_text(aes(label=name)) +
  geom_edge_link(aes(start_cap = label_rect(node1.name), end_cap = label_rect(node2.name)))
```

# 고유벡터 중심성 구하기

```r
JagaEdge %>% as_tbl_graph() %>%
  mutate(eig=centrality_eigen()) %>%
  as_tibble %>% 
  arrange(desc(eig))
```

![CenEig](./photo/Centr_Eigen.PNG)

- 인디 장르의 eig가 1에 수렴함 -> 장르별 관계성이 제일 높다고 볼 수 있음
- 결과값이 다 다르게 나왔으며 순위는 인디, 액션, 어드벤처로 뒤따름
